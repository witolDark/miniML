## Заміна відсутніх значень

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

data_copy = data.copy()
data_copy.columns = data_copy.columns.str.replace('\t', '')

# 1. Заміна відсутніх значень у числових колонках
numeric_cols = ['Previous qualification (grade)', 'Course', 'Daytime/evening attendance', 
                'Previous qualification', 'Mother\'s occupation', 'Father\'s occupation', 'International',
                'Curricular units 1st sem (enrolled)', 'Curricular units 1st sem (approved)', 
                'Curricular units 1st sem (grade)', 'Curricular units 1st sem (without evaluations)', 
                'Curricular units 2nd sem (credited)', 'Curricular units 2nd sem (enrolled)', 
                'Curricular units 2nd sem (evaluations)', 'Curricular units 2nd sem (grade)', 
                'Inflation rate']

imputer_numeric = SimpleImputer(strategy='mean') 
data_copy[numeric_cols] = imputer_numeric.fit_transform(data_copy[numeric_cols])

# 2. Заповнення пропущених значень у категоріальних колонках
categorical_cols = ['Marital status', 'Nacionality', 'Attendance', 'Field of Study', 'Special Needs', 
                    'Citizenship', 'Family Position', 'Target']

imputer_categorical = SimpleImputer(strategy='most_frequent')  # Заповнюємо найчастішими значеннями
data_copy[categorical_cols] = imputer_categorical.fit_transform(data_copy[categorical_cols])
```

## Кількість пропущених значень в оновленому датасеті

```python
missing = []

for col in data_copy.columns:
    missing_count = data_copy[col].isnull().sum()  
    print(f"{col}: {missing_count}")  
    if missing_count != 0:
        missing.append(col)  
```

## Кодування категоріальних ознак (One-Hot Encoding)

```python
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_categorical = encoder.fit_transform(data_copy[categorical_cols])

# Додаємо закодовані категорії до основного датасету
encoded_df = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out())
data_copy = pd.concat([data_copy, encoded_df], axis=1)


# Видаляємо оригінальні категоріальні колонки після кодування
data_copy.drop(columns=categorical_cols, inplace=True)
```

```python
data_copy.info()
```

## Обробка аномальних значень

```python
def handle_outliers(df, column):
    # Визначаємо межі IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Заміна аномалій на медіанне значення
    median_value = df[column].median()
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), median_value, df[column])

# Обробляємо аномалії для вибраних числових колонок
for col in numeric_cols:
    handle_outliers(data_copy, col)
```

## Масштабування числових колонок

```python
# 5. Масштабування числових колонок (за потреби)
scaler = StandardScaler()
data_copy[numeric_cols] = scaler.fit_transform(data_copy[numeric_cols])  # Масштабування всіх числових колонок
```

## Висновок 

У процесі підготовки даних для моделювання були виконані наступні ключові етапи та застосовані методи:

### 1. Копіювання оригінального датасету
- Створено резервну копію оригінального датасету для збереження первинних даних.

### 2. Виправлення назв колонок
- Зайві символи (наприклад, табуляція) були видалені з назв колонок для уникнення помилок під час доступу до даних.

### 3. Обробка пропущених значень
- **Числові колонки:** Використання методу **Mean Imputation** для заміни пропущених значень середнім значенням колонок.
- **Категоріальні колонки:** Використання методу **Most Frequent Imputation** для заміни пропущених значень найбільш поширеними категоріями.

### 4. Кодування категоріальних ознак
- Застосування **One-Hot Encoding** для перетворення категоріальних змінних у числовий формат, що дозволяє використовувати їх в алгоритмах машинного навчання.

### 5. Обробка аномальних значень
- Використання методу **IQR (Interquartile Range)** для виявлення аномалій у числових колонках та їх позначення без заміни на конкретні значення, щоб зберегти розподіл даних.

### 6. Масштабування числових ознак
- Застосування **StandardScaler** для стандартизації даних (приведення до середнього 0 та стандартного відхилення 1), що є критично важливим для алгоритмів, чутливих до масштабу.
