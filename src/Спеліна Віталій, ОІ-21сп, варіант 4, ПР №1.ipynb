## Назви полів даних:

- **unnamed: 0** - невизначене поле, може бути індексом або номером запису
- **marital status** - сімейний статус студента 
- **application mode** - спосіб подачі заявки
- **application order** - порядок подачі заявки
- **course** - курс навчання, на який зарахований студент
- **daytime/evening attendance** - відвідування занять (денна або вечірня форма навчання)
- **previous qualification** - попередня кваліфікація студента (рівень освіти перед вступом)
- **previous qualification (grade)** - оцінка за попередню кваліфікацію
- **nacionality** - національність студента
- **mother''s qualification** - освітній рівень матері студента
- **father''s qualification** - освітній рівень батька студента
- **mother''s occupation** - професія матері студента
- **father''s occupation** - професія батька студента
- **admission grade** - оцінка за вступ
- **displaced** - вказує, чи був студент переміщеним (наприклад, через конфлікт чи інші обставини)
- **educational special needs** - спеціальні освітні потреби студента
- **debtor** - вказує, чи є студент боржником за оплату навчання
- **tuition fees up to date** - вказує, чи оплачені навчальні збори
- **gender** - стать студента
- **scholarship holder** - вказує, чи отримує студент стипендію
- **age at enrollment** - вік студента на момент вступу
- **international** - вказує, чи є студент міжнародним
- **curricular units 1 st sem (credited)** - кількість зарахованих предметів у першому семестрі
- **curricular units 1 st sem (enrolled)** - кількість предметів, на які студент записався у першому семестрі
- **curricular units 1 st sem (evaluations)** - кількість оцінених предметів у першому семестрі
- **curricular units 1 st sem (approved)** - кількість затверджених (успішно зданих) предметів у першому семестрі
- **curricular units 1 st sem (grade)** - середня оцінка за предмети першого семестру
- **curricular units 1 st sem (without evaluations)** - кількість предметів у першому семестрі без оцінювання
- **curricular units 2 nd sem (credited)** - кількість зарахованих предметів у другому семестрі
- **curricular units 2 nd sem (enrolled)** - кількість предметів, на які студент записався у другому семестрі
- **curricular units 2 nd sem (evaluations)** - кількість оцінених предметів у другому семестрі
- **curricular units 2 nd sem (approved)** - кількість затверджених (успішно зданих) предметів у другому семестрі
- **curricular units 2 nd sem (grade)** - середня оцінка за предмети другого семестру
- **curricular units 2 nd sem (without evaluations)** - кількість предметів у другому семестрі без оцінювання
- **unemployment rate** - рівень безробіття у регіоні
- **inflation rate** - рівень інфляції
- **gdp** - валовий внутрішній продукт регіону (економічний показник)
- **target** - мета навчання або статус студента (успішно завершив чи був відрахований)
- **citizenship** - громадянство студента
- **family position** - сімейне положення студента
- **attendance** - відвідуваність студента
- **field of study** - галузь навчання (факультет чи напрямок)
- **special needs** - чи має студент особливі потреби

## Відсоток студентів, які є боржниками, серед різних вікових груп
## Графік ілюструє відсотковий розподіл студентів з боргами по оплаті навчання у різних вікових групах.

```python
ds['Age group'] = pd.cut(ds['Age at enrollment'], bins=[15, 20, 25, 30, 35, 40, 45, 50], labels=['15-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50'])
debtors_by_age = ds[ds['Debtor'] == 1]['Age group'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 8))
debtors_by_age.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis', legend=True)
plt.title('Відсоток студентів-боржників серед різних вікових груп')
plt.ylabel('')
plt.show()
```

## Залежність оцінок студентів від рівня безробіття
## Цей графік відображає взаємозв’язок між середньою оцінкою студентів та рівнем безробіття у регіоні.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Unemployment rate', y='Curricular units 1st sem (grade)', data=ds, hue='Target', alpha=0.6)
plt.title('Залежність оцінок студентів від рівня безробіття')
plt.xlabel('Рівень безробіття')
plt.ylabel('Середня оцінка за 1-й семестр')
plt.show()
```

    Студенти, які навчаються на денній формі (Full-Time), частіше випускаються або залишаються в навчанні порівняно з тими, хто навчається онлайн або на неповній формі. Це може свідчити про те, що більше залучення до навчального процесу сприяє кращим результатам.

4. **Форма навчання та курс**
